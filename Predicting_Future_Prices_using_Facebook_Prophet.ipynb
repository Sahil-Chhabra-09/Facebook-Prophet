## Preparing the data

facebook prophet only needs date and average price, null or missing values are easily handled by it.

```python
avocado_df
```

```python
avocado_prophet_df = avocado_df[['Date', 'AveragePrice']]
```

```python
avocado_prophet_df
```

```python
avocado_prophet_df.columns = ['ds', 'y']
avocado_prophet_df
```

## Understanding Intiution behind Facebook's Prophet

* It is a procedure for forecasting time series data based on an additive model where non-linear trends are fit with yearly, weekly and daily seasonality, plus holdiay effects
* It works best with time series that have strong seasonal effects and several seasons of historical data

* it implements an additive regression model with four elements:
  * A piecewise linear, Prophet automatically picks up change points in the data and identifies any change in trends
  * A yearly seasonal component modeled using Fourier series
  * A weekly seasonal component
  * A holiday list that can be manually provided

* Generalized Additive Model:
  * The purpose of GAM is to maximize the quality of predition of a dependent variable Y from various distributions, by estimating unspecific (non-parametric) functions of the predictor variables which are "connected" to the dependent variable via a link function

### Prophet's features:

* Accurate and Fast:
  * can generate result in seconds
* Automatic:
  * no need to perform data preprocessing
  * works with missing data with several outliers
* Domain Knowledge Integration:
  * users can tweak forecast by manually adding domain specific knowledge

## Developing model and making predictions

```python
model = Prophet()
model.fit(avocado_prophet_df)
```

```python
# forecasting the future
future = model.make_future_dataframe(periods = 365)
forecast = model.predict(future)
```

```python
forecast
```

```python
figure = model.plot(forecast, xlabel = 'Date', ylabel = 'Price')
```

```python
figure2 = model.plot_components(forecast)
```
